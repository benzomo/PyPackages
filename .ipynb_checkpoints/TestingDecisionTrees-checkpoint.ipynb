```python
from Algos.Viz import *
from Algos.ModelsML import ts_LSTM, DecisionTree
from Algos.functions import *

import numpy as np, pandas as pd
import statsmodels.api as sm

import sklearn

from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

from functools import reduce

import seaborn as sns
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
```

```python
ohe = OneHotEncoder()
le = LabelEncoder()

datapath = "/home/benmo/OneDrive/GitHub/Kaggle/HousePrices/data"
test = pd.read_csv(datapath + '/train.csv')
y = test.SalePrice

test.drop(columns='SalePrice',inplace=True)

X_train, y_train = test.copy(), y.copy()

cat = list(filter(lambda x: x not in X_train.corr().columns.tolist(), X_train.columns))
X_train[cat] = X_train[cat].astype(str)


int_labels = dict(zip(cat, list(map(lambda x: le.fit_transform(X_train[x]),
                                        cat))))

onehot_labels = dict(zip(cat, list(map(lambda x: ohe.fit_transform(
        int_labels[x].reshape(-1,1)), cat))))

bobby = [pd.DataFrame(onehot_labels[key].toarray()).rename(
            columns = lambda x: key + str(x)) for key in onehot_labels.keys()]



bobby = reduce(lambda x, y: pd.concat((x, y), axis=1), bobby)

num_cols = list(filter(
    lambda x: x not in cat, X_train.columns))


X_train[num_cols] = X_train[num_cols].fillna(X_train[num_cols].mean().to_dict())
X_train = pd.concat((X_train[num_cols], bobby), axis=1)

X_train, X_test, y_train, y_test = train_test_split(X_train, 
                                                       y_train,
                                                       shuffle=True,
                                                       train_size=0.75)

X_train, y_train, X_test, y_test = map(lambda x: x.reset_index(drop=True), 
                                       (X_train, y_train, X_test, y_test))
```

```python
dt1 = DecisionTree(ttype='generic')
dt2 = DecisionTree(ttype='aboost')

dt1_fit = dt1.fit(X_train,y_train)
dt2_fit = dt2.fit(X_train,y_train)

y_1, y_2 = dt1.predict(X_test), dt2.predict(X_test)
```

```python
pd.concat((y_test,pd.Series(y_1)),axis=1).plot()
pd.concat((y_test,pd.Series(y_2)),axis=1).plot()
```

```python
mean_squared_error(y_test, y_1, multioutput='raw_values')
```
